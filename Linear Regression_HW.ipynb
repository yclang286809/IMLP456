## 用訓練資料來 fit 函數
1. 只用訓練資料集的資料進行linear regression演算法<br>
2. 並計算出訓練階段的MSE
3. 畫出目標點(藍色)與預測點(紅色)的對應圖
![](images/LR_train.PNG)

```python
from sklearn.metrics import mean_squared_error
import numpy as np # 確保 numpy 有被匯入

# 1. 資料維度轉換 (關鍵步驟！)
# train_test_split 切出來的 x_train 是 1D array (40,)
# 但 LinearRegression 訓練時需要 2D array (40, 1)
x_train_reshaped = x_train.reshape(-1, 1)

# 2. 建立與訓練模型
model = LinearRegression()
model.fit(x_train_reshaped, y_train)

# 3. 預測 (為了計算 MSE 與畫圖)
y_train_pred = model.predict(x_train_reshaped)

# 4. 計算訓練階段的 MSE
mse = mean_squared_error(y_train, y_train_pred)
print(f"訓練階段的 MSE: {mse}")

```

```python
# 5. 畫出對應圖
plt.figure(figsize=(10, 6))

# 畫出真實數據 (藍色)
plt.scatter(x_train, y_train, color='blue', label='Actual Data')

# 畫出預測線 (紅色)
# --- 修正點開始 ---
# 使用 .flatten() 把 (N, 1) 的矩陣壓平變成 (N,)，這樣 argsort 才能正確算出從小到大的索引
sorted_idx = np.argsort(x_train.flatten()) 

# 使用這些索引來重新排列 x 和 y，確保線條是順著畫的
plt.plot(x_train[sorted_idx], y_train_pred[sorted_idx], 
         color='red', linewidth=3, label='Prediction Line')
# --- 修正點結束 ---

plt.title(f'Linear Regression Training Result (MSE={mse:.4f})')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## 將訓練出來的函數預測測試集的X值
1. 使用剛剛訓練出來的模型進行測試資料集的資料預測*注意reshape<br>
使用X = 2.44897959，預測出來數值應該為 4.3025375<br>
所有測試集資料
$$\widehat{y}=xw+b=w_{1}x_{1}+b$$

```python
x_test_reshaped = x_test.reshape(-1, 1)
y_test_pred = model.predict(x_test_reshaped)
```

```python
check_val = np.array([[2.44897959]])
pred_val = model.predict(check_val)
print(f"驗證 X=2.44897959 的預測結果: {pred_val[0]:.7f} (目標: 4.3025375)")
print("-" * 30)
```

2. 並計算出測試階段的MSE

```python
mse_test = mean_squared_error(y_test, y_test_pred)
print(f"測試階段的 MSE: {mse_test}")


# 3. 畫出目標點(藍色)與預測點(紅色)的對應圖
plt.figure(figsize=(10, 6))

# 畫出真實答案 (Ground Truth) - 藍色點
plt.scatter(x_test, y_test, color='steelblue', label='Ground Truth', s=70, alpha=0.8)

# 畫出模型預測 (Predict) - 紅色點
plt.scatter(x_test, y_test_pred, color='red', label='Predict', s=70)

# 設定標題
w = model.coef_[0]
b = model.intercept_
plt.title(f'f(x)={w:.1f}x+{b:.1f}, mse:{mse_test:.2f}')

plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
